## 10815 숫자 카드

```python
n = int('5')
nLi = list(map(int, '6 3 2 10 -10'.split()))
m = int('8')
mLi = list(map(int, '10 9 -5 2 3 4 5 -10'.split()))

nLi.sort()
mArr = [0] * m
for i in range(m) :
    mid = nLi[-1] // 2
    midLen = len(nLi) // 2
    if mLi[i] == mid :
        mArr[i] = 1
    elif mLi[i] <= mid :
        for j in range(midLen) :
            j += midLen
            if nLi[j] == mLi[i] :
                mArr[i] = 1
    else :
        for j in range(midLen) :
            j = midLen - j
            if nLi[j] == mLi[i] :
                mArr[i] = 1
print(mArr)
```

### for 문으로 구현

```python
n = int('5')
nLi = list(map(int, '6 3 2 10 -10'.split()))
m = int('8')
mLi = list(map(int, '10 9 -5 2 3 4 5 -10'.split()))

# n = int(input())
# nLi = list(map(int, input().split()))
# m = int(input())
# mLi = list(map(int, input().split()))

nLi.sort()
mArr = [0] * len(mLi)
for i in range(len(mLi)) :
    left, right = 0, len(nLi) - 1
    mid = (left+right) // 2
    targetIndex = mid
    already = 0
    # print(i)
    for _ in range(len(mLi)) :
        # print('target', targetIndex, _)
        if targetIndex == right or targetIndex == 0 :            
            already += 1
        if already == 2 :
            # print('break', _)
            break
                
        if nLi[targetIndex] == mLi[i] :
            mArr[i] = 1
            # print('=1')
            break
        elif nLi[targetIndex] < mLi[i] :
            # print('+1')
            targetIndex += 1
        elif nLi[targetIndex] > mLi[i] :
            # print('-1')
            targetIndex -= 1
        
a = ''
for i in mArr :
    a += (str(i)+' ')
print(a)
```

### while 문 (gpt)

```python
def binary_search(cards, target):
    left, right = 0, len(cards) - 1

    while left <= right:
        mid = (left + right) // 2
        if cards[mid] == target:
            return 1  # 숫자가 존재함
        elif cards[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return 0  # 숫자가 존재하지 않음

def main():
    # 입력 처리
    # n = int(input())
    # cards = list(map(int, input().split()))
    # m = int(input())
    # targets = list(map(int, input().split()))

    n = int('5')
    cards = list(map(int, '6 3 2 10 -10'.split()))
    m = int('8')
    targets = list(map(int, '10 9 -5 2 3 4 5 -10'.split()))

    # 입력받은 카드를 정렬
    cards.sort()

    # 각 숫자 카드에 대해 이분 탐색을 수행하여 결과 출력
    for target in targets:
        result = binary_search(cards, target)
        print(result, end=' ')

if __name__ == "__main__":
    main()
```
